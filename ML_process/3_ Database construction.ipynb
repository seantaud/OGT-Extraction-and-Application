# 1、OGT data processing

## Read data

### Model test experimental data set

```python
def get_QA():
    import pandas as pd

    OGT_QA = pd.read_excel('dataset/OGT_QA.xlsx')
    OGT_QA = OGT_QA[['taxonomy_id', "OGT", 'context']]

    def isFloat(x):
        try:
            float(x)
            return True
        except:
            return False

    def data_standardization(OGT_QA):
        ogt_list = []
        for each in OGT_QA.values:
            temp = each[-2]
            if isFloat(temp):
                each[1] = float(temp)
                ogt_list.append(each)
                continue

            if '-' in temp:
                min = temp.split('-')[0]
                max = temp.split('-')[-1]
                mean = (float(min) + float(max))/2
                each[1] = float(mean)
                ogt_list.append(each)
                continue

            if '±' in temp:
                mean = float(temp.split('±')[0])
                each[1] = float(mean)
                ogt_list.append(each)
                continue

            if 'K' in temp:
                mean = float(temp.split('K')[0]) - 272.15
                each[1] = float(mean)
                ogt_list.append(each)
                continue

        return pd.DataFrame(ogt_list, columns = ['taxonomy_id', 'OGT_journal', 'context'])

    OGT_QA = data_standardization(OGT_QA)

    return OGT_QA

OGT_QA = get_QA()
```

## context deduplication

```python
def fully_drop(data):    
    import pandas as pd
    import difflib

    # Create a function to calculate the similarity
    def calculate_similarity(str1, str2):
        matcher = difflib.SequenceMatcher(None, str1, str2)
        return matcher.ratio()

    # Grouping according to taxonomy and OGT
    groups = data.groupby(['taxonomy_id', 'OGT_journal'])

    # Empty data box for storing results
    result = pd.DataFrame(columns=data.columns)

    # Walk through each group
    for group_name, group_data in groups:
        if len(group_data) > 1:
            most_similar = None
            max_similarity = 0.0

            # Iterate over the data within the group, looking for the most similar context
            for i, row1 in group_data.iterrows():
                for j, row2 in group_data.iterrows():
                    if i != j:
                        similarity = calculate_similarity(row1['context'], row2['context'])
                        if similarity > max_similarity:
                            max_similarity = similarity
                            most_similar = row1 if len(row1['context']) <= len(row2['context']) else row2

            # Add the most similar data to the result data box
            result = result.append(most_similar)
        else:
            result = result.append(group_data)
    
    return result
```

```python
Search_exp_dataset = fully_drop(OGT_QA)
Comparation_exp_dataset = fully_drop(OGT_journal)
```

## Data cleaning

### The test data set is treated with standard deviation

```python
def clean_std(OGT_journal):
    OGT_journal['Doft_value'] = OGT_journal.groupby('taxonomy_id')['OGT_journal'].transform('std')
    OGT_journal = OGT_journal.fillna(value=0)
    
    tax = OGT_journal['taxonomy_id'].value_counts()
    dfall = None

    for each in tax.index:
        df = OGT_journal.loc[OGT_journal.taxonomy_id == each]
        df = df.loc[df.Doft_value <=5]
        if dfall is None: 
            dfall = df.copy()
        else: 
            dfall = pd.concat([dfall,df], axis=0)
    
    dfall = dfall[['taxonomy_id', 'OGT_journal', 'context']]
    print(dfall.shape)
    return dfall

Search_exp_dataset_cleaned = clean_std(Search_exp_dataset)
```

### The two data were averaged, and the retained biological energy corresponded to their respective OGT and culture temperature one by one, which involved temperature de-duplication operation

```python
def load_OGT_cultivation():
    OGT_cultivation = pd.read_csv('dataset/OGT_cultivation.tsv', sep = '\t')
    OGT_cultivation = OGT_cultivation[['taxid', 'organism', 'temperature']]
    OGT_cultivation = OGT_cultivation.rename(columns  = {'organism':'scientific_name',
                                                   'taxid':'taxonomy_id',
                                                   'temperature':'OGT_cultivation'})
    
    return OGT_cultivation

def merge_ogt(data):
    OGT_cultivation = load_OGT_cultivation()
    OGT_cultivation = OGT_cultivation.drop_duplicates()
    OGT_cultivation = OGT_cultivation.groupby('taxonomy_id')['OGT_cultivation'].mean().reset_index()
    
    taxid_common = []
    data['OGT_cultivation'] = None
    for x in data.values:
        for y in OGT_cultivation.values:
            if int(x[0]) == int(y[0]):
                x[3] = y[1]
                taxid_common.append(x)
                break

    OGT_final = pd.DataFrame(taxid_common, columns = ['taxonomy_id', 'OGT_journal', 'context',  'OGT_cultivation'])
    
    OGT_final = OGT_final.drop_duplicates(subset=['taxonomy_id', 'OGT_journal'])

    # Average by the 'taxonomy_id' and 'OGT_journal' columns while retaining the other column data
    result = OGT_final.groupby(['taxonomy_id']).agg({
        'context': 'first',    
        'OGT_cultivation': 'first',    
        'OGT_journal': 'mean'       
    }).reset_index()

    return result

Search_exp_final_data = merge_ogt(Search_exp_dataset_cleaned)
Comparation_exp_final_data = merge_ogt(Comparation_exp_dataset[['taxonomy_id', 'OGT_journal', 'context']])
print(len(Search_exp_final_data), len(Comparation_exp_final_data))
```
